```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df= pd.read_csv("../python-ml-course-master/fires-all.csv")
```

```python
df_comp=df.copy()
```

```python
df_comp.head()
```

```python
df_comp.tail()
```

```python
#Ordeno por fechas
df_comp=df_comp.sort_values(by='fecha')
df_comp= df.reset_index(drop=True)
df_comp
```

```python
#Filtro latitudes que no corresponde a Peninsula Iberica:
df_comp= df_comp.drop(df_comp[df_comp['lat']>44.0].index)
df_comp= df_comp.reset_index(drop=True)
df_comp= df_comp.drop(df_comp[df_comp['lat']<36.0].index)
df_comp= df_comp.reset_index(drop=True)
```

```python
df_comp.tail()
```

```python
x=df_comp["lat"].tolist()
y=df_comp["superficie"].tolist()
plt.plot(x,y,"ro")
plt.title("Superficie por incendio vs latitud")
plt.show()
```

```python
#Redondeo latitudes para posteriormente hacer agrupaciones por superficie,area:
lat_round=[]
for i in range(len (df_comp)):
    df_comp["lat"][i]=round(df_comp["lat"][i])
    lat_round.append(df_comp["lat"][i])
```

```python
df_comp
```

```python
#Transformo a categorical las latitudes redondeadas:
df_comp["lat"]=pd.Categorical(df_comp["lat"])
```

```python
#Examino por numero de incendios en el total de aÃ±os en los grupos e latitud:
tab_Fabs= pd.crosstab(index=df_comp["lat"],
                columns= "incendios totales 2001-2015")

tab_Fabs
```

```python
tab_Fabs.plot(kind="bar",title="Frecuencia absoluta de incendios(2001-2015)",legend=False)
```

```python
#Agrupo por latitud para ver la superficie total:
df_lat=df_comp.groupby("lat").sum()
df_lat
```

```python
df_lat["superficie"].plot(kind="bar",title="Superficie total afectada por latitud")
```

```python
len(df_comp)
```

```python
dt=[]
for i in range(len(df_comp)):
    dt.append(datetime.strptime(df["fecha"][i], '%Y-%m-%d'))
df_comp["fecha"]=dt
```

```python
df_comp["fecha"].describe()
```

```python
#Agrupo por fecha y latitud para estudiar las series temporales por latitud:
df_comp=df_comp.groupby(["fecha","lat"],as_index=False).sum()
df_comp
```

```python
#Separo el dataframe por grupos de latitud:
df36=df_comp[::9]
df36= df36.reset_index(drop=True)
df37=df_comp[1::9]
df37= df37.reset_index(drop=True)
df38=df_comp[2::9]
df38= df38.reset_index(drop=True)
df39=df_comp[3::9]
df39= df39.reset_index(drop=True)
df40=df_comp[4::9]
df40= df40.reset_index(drop=True)
df41=df_comp[5::9]
df41= df41.reset_index(drop=True)
df42=df_comp[6::9]
df42= df42.reset_index(drop=True)
df43=df_comp[7::9]
df43= df43.reset_index(drop=True)
```
